```python
import tensorflow as tf
import numpy as np
import matplotlib. as plt
import pandas as pd
```

```python
print('Tensorflow Version :{}'.format(tf.__version__))
```

```python
data = np.random.rand(100)
```

```python
target = 0.3*data + 0.1
```

```python
data = pd.DataFrame(data,columns = ['Education'])
data
```

```python
target = pd.DataFrame(target,columns = ['Income'])
target
```

```python
data = pd.concat([data,target],axis = 1)
```

```python
data.head()
```

```python
x = data.Education
y = data.Income
```

```python
model = tf.keras.Sequential()#model = tf.keras.Sequential(tf.keras.layers.Dense(units=1,input_shape=[1]))[1]或者元组（1，）b
```

tf.keras.Sequential()表示的是按照层数顺序建立神经网络模型 Dense中units表示的是神经网络的层数 input_shape表示输入数据的维度

```python
model.add(tf.keras.layers.Dense(units=1,input_shape=[1]))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='mse')
```

```python
history = model.fit(x,y,epochs=500)
```
